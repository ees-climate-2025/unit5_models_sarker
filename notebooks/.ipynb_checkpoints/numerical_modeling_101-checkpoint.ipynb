## What is a numerical model?

A __numerical model__ is a type of mathematical model, where __numerical methods__ are used to approximate the solutions to the mathematical equations - which may be difficult or impossible to find analytical (exact) solutions for.

> A numerical model breaks down mathematical equations into __discrete parts (discretization)__ and then uses computers to iterate across these discrete units to find approximate solutions to the equations.


> ##### Examples of numerical models include weather forecasting models, climate models, atmospheric models, and ocean-sea ice models, which we will be discussing and using in this course. 

---

This notebook walks through the general concepts of building a simple numerical model that starts from some set of __initial conditions__, steps forward in time, and updates the values (__interates__). We will use the basic concepts learned here in another notebook where we solve for the temperature of Earth's surface.

In this case, we are applying __Newton's Law of Cooling__ which states that the temperature of a body changes at a rate proportional to the _difference in temperature_ between the body and its surroundings. This means that a hot beverage (e.g., tea or coffee) will cool down at a faster rate in a cold room, compared to a room at a comfortable temperature.

This law takes the form of a __differential equation__ (an equation that contains derivatives of a function), specifically an __ordinary differential equation (ODE)__, with only one independent variable (time):

### $\frac{dT}{dt} = -k (T - T_{\text{ambient}})$

where _T_ is the Temperature of the body (coffee, soda, tea, etc), __$T_{\text{ambient}}$__ is the temperature of the surrounding environment, _k_, is a positive constant, and $\frac{dT}{dt}$ is the "incremental change in temperature" over an "incremental change in time".

__There for a greater difference in temperature between the body, T, and __$T_{\text{ambient}}$__, the greater $\frac{dT}{dt}$.__

We can translate this ODE into Python and write a simple model that __steps forward in time ($dt$)__ and solves for an objects temperature at a given time (_iterates_). According to Newton's Law of Cooling, the following parameters must be set as initial conditions / constants:

* __Intial temperature of the body__ (T; degrees C)
* __The temperature of the room__ (__$T_{\text{ambient}}$__; degrees C)
* __Cooling rate__ (k; specified as a cooling constant, degrees C per minute) (k)

We then need to specify the __time step__ for the model and the __simulation length__:
* __dt__ (time step in minutes)
* __run length__ (time in minutes)

#### Equation 1:

## $$ 
\frac{dT}{dt} = -k (T - T_{\text{ambient}})
$$

#### We can write this equation in discrete form and solve it to get the numerical solution. For the sake of understanding lets write this out as a series of steps, first translating the equation above into an algebraic form:

## $$
\frac {\Delta T}{\Delta t} = -k (T - T_{\text{ambient}})
$$

## $$
{\Delta T} = -k (T - T_{\text{ambient}}){\Delta t}
$$

## $$
{T_{n+1}} - {T_{n}}  = -k (T - T_{\text{ambient}}){\Delta t}
$$

#### Equation 2:

## $$
T_{n+1} = T_n - k (T_n - T_{\text{ambient}}) \, \Delta t
$$

#### In Equation 2 above, $T_{n+1}$ represents the temperature of the coffee at the "next point in time" and $T_{n}$ is the coffee temperature at the "current point in time". The $\Delta t$ is the "discretized" time.

### __Thus, we are solving for the next temperature using the _current conditions___. In more modeling words, we are __approximating the derivative in Equation 1 by a _forward difference_.

### The foward differencing method to approximate a solution to our differential equation above is known as the __Euler method__ to step the numerical solution forward in time.

### We don't have a model yet, all we've done is set up the initial conditions and model parameters. For this to be a numerical model, we need it to solve the differential equation by computing Temperature for a given change in time (dt):

Here we write a simple _while loop_ that iterates in time to accomplish this task that says:

> "While time is less than the length of the run (run_length), compute the incremental change in temperature (dt) of the coffee and add this to the initial temperature (temp), then update the coffee's temperature (temp) and the time value (dt) and print out the new temperature"

The first line of the while loop solves for dT by rearranging Equation 1 to:

## $$
{dT} = -k (T - T_{\text{ambient}})* dt
$$

#### Or written in algebraic / discrete form:

## $$
{\Delta T} = -k (T - T_{\text{ambient}})* {\Delta t}
$$


```python
# Initialize values
initial_temp = 90   # Initial temperature of coffee in °C
ambient_temp = 22   # Room temperature in °C
cooling_rate = 0.1  # Cooling rate constant (per minute)
dt = 1              # Time step in minutes
run_length = 60     # Total simulation time in minutes

# Step forward in time from initial values
temp = initial_temp
time = 0

while time < run_length:
    # Compute temperature change
    dT = -cooling_rate * (temp - ambient_temp) * dt
    
    # Update temperature
    # Note that this can be written in short hand as: #temp += dT
    temp = temp + dT
    
    # Update time
    # Note that this can be written in short hand as: #time += dt
    time = time + dt
   
    # Print current time and temperature
    print(f"{time}\t\t{temp:.2f}")
```

### Great so now instead of printing the values, lets place them inside of a list so that we can plot the rate of temperature change: 

```python
# Initialize values
initial_temp = 90   # Initial temperature of coffee in °C
ambient_temp = 22   # Room temperature in °C
cooling_rate = 0.1  # Cooling rate constant (per minute)
dt = 1              # Time step in minutes
run_length = 60     # Total simulation time in minutes

# Initialize variables
temp = initial_temp
time = 0

# Create empty lists to store the results:
times = [time]
temps = [temp]

# Step forward in time
while time < run_length:
    # Compute temperature change
    dT = -cooling_rate * (temp - ambient_temp) * dt
    
    # Update temperature
    temp += dT
    # Update time
    time += dt
    
    # Store results in the lists that I initialized above outside of the loop
    times.append(time)
    temps.append(temp)
```

### Plot this

```python
# Plotting
plt.figure(figsize=(8, 5))

# create a plot with x = time, y  = temp
plt.plot(times, temps, marker='o', linestyle='-', color='teal')
plt.axhline(ambient_temp, color='gray', linestyle='--', label='Ambient Temperature')
plt.title("Coffee Cooling Over Time")
plt.xlabel("Time (minutes)")
plt.ylabel("Temperature (°C)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

### Now lets write this model into a function where a user can specify the coffee temperature, the room temperature, the time step, and the run length:

```python
def newton_cooling(t_body, t_ambient, dt, run_length):
    
    """
    Simulates Newton's Law of Cooling using the forward Euler method.

    This function models the cooling of an object (e.g., a cup of coffee) over time as it approaches
    the ambient temperature. It uses a discrete time-stepping approach to approximate the temperature
    evolution based on Newton’s law of cooling:

        dT/dt = -k * (T - T_ambient)

    Parameters:
    ----------
    t_body : float
        Initial temperature of the object (°C).
    t_ambient : float
        Ambient (surrounding) temperature (°C).
    dt : float
        Time step size (minutes).
    run_length : float
        Total simulation time (minutes).

    Returns:
    -------
    times : list of float
        List of time values at each step.
    temps : list of float
        Corresponding list of temperature values over time.

    Example:
    -------
    times, temps = newton_cooling(90, 22, 1, 60)
    """
    
    temp = t_body
    time = 0
    cooling_rate = 0.1
    
    times = [time]
    temps = [temp]
    
    # Step forward in time
    while time < run_length:
        # Compute temperature change
        dT = -cooling_rate * (temp - t_ambient) * dt
    
        # Update temperature
        temp += dT
        # Update time
        time += dt
    
        # Store results in the lists that I initialized above outside of the loop
        times.append(time)
        temps.append(temp)
        
    return times, temps
```

## $$
T_{n+1} = T_n - k (T_n - T_{\text{ambient}}) \, \Delta t

### Now lets write this model into a function where a user can specify the coffee temperature, the room temperature, and the time step:

```python
def newton_cooling(T0, T_ambient, k, dt, t_final):
    """
    Simulates Newton's Law of Cooling using the forward Euler method.

    This function models the cooling of an object (e.g., a cup of coffee) over time as it approaches
    the ambient temperature. It uses a discrete time-stepping approach to approximate the temperature
    evolution based on Newton’s law of cooling in the following form:

        T_{n+1} = T_n - k * (T_n - T_ambient) * dt

    Parameters
    ----------
    T_0 : float
        Initial temperature of the object.
    T_ambient : float
        Ambient (surrounding) temperature.
    k : float
        Cooling constant (1/time units; minutes).
    dt : float
        Time step size.
    t_final : float
        Total simulation time in minutes.

    Returns
    -------
    times : list of float
        Time points for each step.
    temps : list of float
        Temperature values at each time point.
    """
    
    # Initialize lists for time and temperature
    times = [0.0]
    temps = [T_0]
    
    # Number of time steps
    n_steps = int(t_final / dt)
    
    # Iteratively apply the discrete equation
    T = T0
    for n in range(n_steps):
        T = T - k * (T - T_ambient) * dt
        temps.append(T)
        times.append((n + 1) * dt)

    return times, temps
```

```python
# Example parameters
T_0 = 100         # initial temperature
T_ambient = 20    # ambient temperature
k = 0.1          # cooling constant
dt = 0.5          # time step
t_final = 60       # total time

# Run the model
times, temps = newton_cooling(T_0, T_ambient, k, dt, t_final)

# Print the first few results
for t, T in zip(times[:6], temps[:6]):
    print(f"t = {t:.1f} s,  T = {T:.2f} °C")
```

```python
plt.plot(times, temps, marker='o')
plt.xlabel('Time (min)')
plt.ylabel('Temperature (°C)')
plt.title("Coffee Cooling Over Time")
plt.grid(True)
plt.show()
```
