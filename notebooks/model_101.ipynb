## What is a model?

You have probably encountered the word "model" in several contexts in your experience as a student, but _what does one actually mean when they say the word “model”?_

Fundamentally, __a model is a representation of some system__. The first point to recognize is that the use of the word "model" implies that you are not in reality. __Models are approximations / estimates of some real-world system__ and their construction is motivated by real-world observations and processes, however no model can perfectly represent the real-world as they are built by making approximations and assumptions of behavior (e.g., simplication of complex real world processes). However, models are useful tools for providing insight into processes / mechanisms and making predictions / projections that have real-world implications. Weather models, for example, are _numerical models_ that we interact with nearly every single day as we make decisions on what to wear today or three days from now.

## Types of "models"

### Physical Models:
> Physical models are __tangible__ representations of a real-world system / process, often used for teaching purposes. Large-scale features/processes (e.g., ocean circulation) may be shrunk down, or small-scale features/processes (e.g., molecular structures) may be enlarged. These models are useful for developing an understanding of processes operating at scales that are hard to measure or translate to a lab setting.

> __Example__:  __Scientific Solar System Models__ can be used to understand relationships between the Sun and other planets. It is easier to teach students about these topics when they have a model that they can visualize and manipulate right in front of them.

### Conceptual Models

> A conceptual model is a qualitative model (diagram, flow chart, etc) that represents a complex system often using visualizations. This type of model is often very simplified and static (not tangible like a physical model). Conceptual models are built from real-world observations and processes that are distilled down into simplified and organized visualizations or words that provide for easier communication of complex concepts. Conceptual models often serve as an important framework when building and motivating more complex numerical models.

> __Example:__ The Carbon Cycle model. This conceptual model shows the flow of carbon through the Earth System and its carbon reservoirs. The numbers on the figure are from real-world quantitative estimates. One can look at this diagram and get an understanding of the role of each reservoir in the global carbon budget.

### Mathematical Model

> A mathematical model explains a physical process / system in __mathematical equations__ (rather than in visual maps / concepts or tangible objects. These equations may be solved __analytically__ (i.e., with an exact solution) or __numerically__ (a numerical or computational model).

> __Example:__ linear-growth model. The equation for a line is $y = mx + b$. 
Where $y$ is the dependent variable, $x$ is the independent variable (time), $m$ is the rate of change (slope), and b is the y-intercept. The linear growth model allows one to make a prediction of future behavior based on the assumption of __linearity__, e.g., that the rate of change does not change over time (the slope, $m$, is constant). So if one knows the starting value (the y intercept, $b$) and the current rate of change ($m$), the future behavior at time $x$ can be predicted. 

> The growth over time for a linear system is __additive__ .... e.g., if I have a bank account that I am adding 50 dollars into each month, ($m$ = 50 dollars / month) and I started with 100 dollars, after 1 year (12 months), I will have $700.

```python
import numpy as np
import matplotlib.pyplot as plt

slope = 2                 # specify the slope of the line 
b = 10                    # specify the y-intercept
x = np.arange(0,5)        # specify the x values (independent), this could be time in the case of a growth model

y = slope*x + 2           # solve for the dependent variable, y
```

### Plot our linear growth model
### $y = 2x + 10$

```python
plt.plot(x,y,marker='o',label='y = 2x + 10')
plt.title('Plot of Linear Equation: y = 2x + 10')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

> __Example:__ Exponential growth model. The exponential growth model is one where the __rate of growth becomes more rapid proportional to the _current_ total size or number of the population__. In the case of exponential growth, there are greater increases over the course of time .... resulting in a growth curve that looks like an exponential function. The opposite is true for exponential decay.

> The exponential growth model for population growth (assuming there is nothing limiting growth) is: 
> ### $P (t) = P_0 \, e^{kt}$,

> where 
> * $P(t)$ is the population
> * $P_0$ is the initial population
> * $e$ is Euler's number (~2.71828)
> * $t$ is time
> * $k$ is the growth rate.

```python
P_0 = 100                           # initial population
k = 0.2                             # define growth rate 
time = np.linspace(0,20,100)         # time from 0 to 20 years, creates a numpy array

P = P_0 * np.exp(k * time)  # compute the population for each time
```

```python
P
```

```python
plt.plot(time,P,marker='o',label='$P (t) = 100 \, e^{0.2t}$')
plt.title('Plot of exponential equation: $P (t) = P_0 \, e^{kt}$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

### Numerical Models

A __numerical model__ is a type of mathematical model, where __numerical methods__ are used to approximate the solutions to the mathematical equations - which may be difficult or impossible to find analytical (exact) solutions for.

> A numerical model breaks down mathematical equations into __discrete parts (discretization)__ and then uses computers to iterate across these discrete units to find approximate solutions to the equations.


> ##### Examples of numerical models include weather forecasting models, climate models, atmospheric models, and ocean-sea ice models, which we will be discussing and using in this course.
